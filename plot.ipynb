```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
feather = pd.read_csv("results/feather.csv")

plt.plot(feather['size (Mb)'], feather['write time (s)']*1000, label='write time')
plt.plot(feather['size (Mb)'], feather['read time (s)']*1000, label='read time')
plt.legend(['write', 'read'])
plt.xlabel('size (Mb)')
plt.ylabel('time (ms)')
plt.title("Feather file write and read times")
plt.show()
```

```python
csvgz = pd.read_csv("results/csvgz.csv")

plt.scatter(csvgz['size (Mb)'], csvgz['write time (s)']*1000, label='write time')
plt.scatter(csvgz['size (Mb)'], csvgz['read time (s)']*1000, label='read time')
plt.legend(['write', 'read'])
plt.xlabel('size (Mb)')
plt.ylabel('time (ms)')
plt.title("csv.gz file write and read times")
plt.show()
```

```python
# comparison
plt.plot(feather['size (Mb)'], feather['write time (s)']*1000, label='feather write time')
plt.plot(feather['size (Mb)'], feather['read time (s)']*1000, label='feather read time')
plt.plot(csvgz['size (Mb)'], csvgz['write time (s)']*1000, label='csv.gz write time')
plt.plot(csvgz['size (Mb)'], csvgz['read time (s)']*1000, label='csv.gz read time')
plt.yscale('log', base=2)
plt.legend(['feather write', 'feather read', 'csv.gz write', 'csv.gz read'])
plt.xlabel('size (Mb)')
plt.ylabel('time (ms)')
plt.title("Csv.gz vs Feather file write and read times")
plt.show()
```
